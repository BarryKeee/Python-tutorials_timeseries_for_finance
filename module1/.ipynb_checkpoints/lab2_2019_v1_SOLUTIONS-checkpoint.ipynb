## ARMA($p$,$q$) models

**Recall:** a stochastic process process $\{X_t\}$ is an ARMA($p$,$q$) process (autoregressive moving average process of order $p$ and $q$, respectively) if we have
$$
X_t = \varepsilon_{t} + \sum_{i=1}^{p} \phi_i X_{t-i} + \sum_{i=0}^{q} \theta_i \varepsilon_{t-i}
$$
where $\varepsilon_{t}$ is white noise, i.e. $E[\varepsilon_{t}]=0$ and $Var[\varepsilon_{t}]=\sigma^2$

If the coefficients $\phi_i \equiv 0$, then the ARMA($p$,$q$) process collapses to an MA($q$) process. Similarly, if $\theta_i\equiv 0$ then the ARMA($p$,$q$) process collapses to an AR($p$) process.

A slightly more general formulation also used is given by the expression
$$
\sum_{i=0}^{p} \phi_i X_{t-i} = \sum_{i=0}^{q} \theta_i \varepsilon_{t-i}
$$
Note that this formulation introduces $\phi_0,\theta_0$ terms which were implicitly defined to be 1 above. Further, the sign of the AR coefficients is flipped from the first formulation. This is frequently how statistical computing packages treat ARMA($p$,$q$) processes.


Let's simulate some ARMA($p$,$q$) processes and examine their autocorrelation functions (i.e. ACFs).


### Exercise 1: Simulate ARMA($p$,$q$) models

**Problem:** Write a function that simulates an ARMA($p$,$q$) process given parameter inputs.

**Note:** `statsmodels` has a function that generates ARMA($p$,$q$) processes, but for now, write your own function.

```python
### For simulating ARMA processes, specifying the roots can be more informative than coefficients.
def _roots2coef(roots):
    """Given roots, get the coefficients"""
    ### sympy: package for symbolic computation
    from sympy import symbols, expand
    N_roots = len(roots)
    L = symbols("L")
    expr = expand(1)
    for r in roots:
        expr*= (r - L)
    expr = expand(expr)
    coef_list = [expr.coeff(L,n) for n in range(N_roots+1)]
    ### convert to numpy floats
    coefs = np.array(coef_list ).astype(float)
    ### normalize zero lag to 1
    coefs /= coefs[0]
    return coefs

def arma_from_roots(ar_roots=[], ma_roots=[]):
    """Create an ARMA model class from roots"""
    ar_coef = _roots2coef(ar_roots)
    ma_coef = _roots2coef(ma_roots)
    arma_process = sm.tsa.ArmaProcess(ar_coef, ma_coef)
    ### Note: arma_process' has many helpful methods: arcoefs, macoefs, generate_sample, ...
    return arma_process
```

```python
def simulate_arma(ar=[], ma=[], nsample=100, burnin=0):
    """Simulate ARMA data
       Assumption: White noise shocks are Gaussian.
    """
    ### numpy arrays are reversed for easier indexing:
    ar, ma = np.array(ar[::-1]), np.array(ma[::-1])
    ### Orders (does not include zero lag)
    p, q = len(ar), len(ma)
    max_order = max(p,q)
    
    ### Total number of simulations
    Nsim = nsample + burnin
    ### Guassian shocks
    eps = np.random.randn(Nsim)
    ### Initialize t < 0 with zeros
    eps = np.r_[np.zeros(max_order), eps]
    y = np.zeros(Nsim + max_order)
    ### Loop
    for tt in range(max_order, Nsim + max_order):
        y[tt] = np.sum(y[tt-p:tt]*ar) + np.sum(eps[tt-q:tt]*ma) + eps[tt]
    ### Drop initial zeros and burnin
    y = y[max_order + burnin:]
    return y
```

```python
### Test the function
np.random.seed(1)

### 1. Just white noise
y_wn = simulate_arma(nsample=1000)

### 2. AR(1) with high persistence
y_ar1 = simulate_arma(ar=[.75], nsample=1000)

### 3. MA(1)
y_ma1 = simulate_arma(ma=[.5], nsample=1000)

### 4. Non-stationary AR(1) 
y_rw = simulate_arma(ar=[1], nsample=1000)


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,12))
axes[0,0].plot(y_wn)
axes[0,0].set_title('White Noise')

axes[0,1].plot(y_ar1)
axes[0,1].set_title('AR(1)')

axes[1,0].plot(y_ma1)
axes[1,0].set_title('MA(1)')

axes[1,1].plot(y_rw)
axes[1,1].set_title('Random Walk')

fig.tight_layout()
```

### Exercise 2: OLS Estimation

**Problem:** Write a function that returns the parameter estimates and residuals from a simple OLS regression.

**Note:** It's helpful to include an option that deals with adding a constant within the function, although this is up to you. If you do add a constant within your function, it can be useful to return the new matrix as well. 

Once again, `statsmodels` and other Python libraries have functions for running linear regressions. For now, write your own function.

```python
def _sm_calc_ols(y,x, addcon=True):
    """Wrapper for statsmodels OLS regression"""
    X = sm.add_constant(x) if addcon else x
    ols_results = sm.OLS(y,X).fit()
    beta_hat = ols_results.params
    resids = ols_results.resid
    return beta_hat, resids, X
```

```python
def calc_ols(y,x, addcon=True):
    """Calculate OLS coefficients"""
    Nobs = y.shape[0]
    X = np.c_[ np.ones( (Nobs,1) ) , x] if addcon else x
    XX = np.dot(X.T, X)
    Xy = np.dot(X.T,y)
    beta_hat = np.linalg.solve(XX, Xy)
    resids = y - np.dot(X, beta_hat)
    return beta_hat, resids, X
```

### Exercise 3: Estimate the  *ACF* and *PACF* by OLS

Using your OLS function, write functions that both estimate and plot the *ACF* and *PACF* by OLS.

**Note:** compare with the `statsmodels` methods:
* `sm.tsa.stattools.acf`, `sm.graphics.tsa.plot_acf`
* `sm.tsa.stattools.pacf`, `sm.graphics.tsa.plot_pacf`

Both of these functions have multiple methods for estimation; we'll stick with OLS.

```python
def lag_mat(y,nlags, fill_vals=np.nan):
    """Create a matrix of lags of a given vector"""
    y_lags = np.empty((y.shape[0], nlags+1)) #https://docs.scipy.org/doc/numpy/reference/generated/numpy.empty.html
    y_lags.fill(fill_vals)
    ### Include 0 lag
    for lag in range(nlags + 1):
        y_lags[lag:,lag] = np.roll(y, shift=lag)[lag:] #https://docs.scipy.org/doc/numpy/reference/generated/numpy.roll.html
    return y_lags


def calc_acf_lag_ols(y_lags, lag):
    """ACF for a given lag (OLS)"""
    if lag==0: return 1.
    lhs = y_lags[lag:,0]
    rhs = y_lags[lag:,lag:lag+1]
    beta_hat, _,_  = calc_ols(lhs,rhs, addcon=True)
    return beta_hat[-1]

def calc_pacf_lag_ols(y_lags, lag):
    """PACF for a given lag (OLS)"""
    if lag==0: return 1.
    lhs = y_lags[lag:,0]
    rhs = y_lags[lag:,1:lag+1] # need y_lags[lag:,1:lag+1] instead of y_lags[lag:,lag:lag+1]
    beta_hat, _,_  = calc_ols(lhs,rhs, addcon=True)
    return beta_hat[-1]

def calc_acf_ols(y, nlags):
    """ACF for multiple lags"""
    y_lags = lag_mat(y,nlags)
    acf_list = [calc_acf_lag_ols(y_lags,lag) for lag in range(nlags+1) ]
    return np.array(acf_list)

def calc_pacf_ols(y, nlags):
    """PACF for multiple lags"""
    y_lags = lag_mat(y,nlags)
    pacf_list = [calc_pacf_lag_ols(y_lags,lag) for lag in range(nlags+1) ]
    return np.array(pacf_list)

def plot_acf(y, nlags=10, ax=None):
    """Plotting ACF with approx SEs."""
    T = y.shape[0]
    ### approx SEs
    se_approx = 1/np.sqrt(T)
    ### set up figure
    if ax is None:
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))
    ### ACF
    ax.plot(calc_acf_ols(y, nlags))
    ax.axhline(1.96*se_approx)
    ax.axhline(-1.96*se_approx)
    ax.set_title('ACF')
    
def plot_pacf(y, nlags=10, ax=None):
    """Plotting PACF with approx SEs."""
    T = y.shape[0]
    # approx SEs
    se_approx = 1/np.sqrt(T)
    # set up figure
    if ax is None:
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))
    # PACF
    ax.plot(calc_pacf_ols(y, nlags))
    ax.axhline(1.96*se_approx)
    ax.axhline(-1.96*se_approx)
    ax.set_title('PACF')    
```

```python
### Test the functions 

### Simulate some data
np.random.seed(2)

### ARMA(1,1) process
y_arma11 = simulate_arma(ar=[.9], ma=[.5], nsample=1000)

nlags = 10
print(calc_acf_ols(y_arma11, nlags))
print(calc_pacf_ols(y_arma11,nlags ))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
plot_acf(y_arma11, nlags , ax=axes[0])
plot_pacf(y_arma11, nlags , ax=axes[1])

fig.tight_layout()
```

### Exercise 6: Diagnostic tests

Given the residuals of an estimated ARMA($p$,$q$) model, write Python functions that calculate the following:
* **Durbin-Watson** statistic
* **Breusch-Pagan** statistic and p-values
* **AIC** and **BIC**

```python
### https://en.wikipedia.org/wiki/Durbin%E2%80%93Watson_statistic
def durbin_watson_stat(resids):
    """Compute Durbin-Watson statistic"""
    numer = (resids[1:] - resids[:-1])**2 # NOTE: Think of resids[:-n] as the same as resids[:len(resids)-n].
    denom = resids**2
    return np.sum(numer)/np.sum(denom)

### https://en.wikipedia.org/wiki/Breusch%E2%80%93Pagan_test
def breusch_pagan_godfrey_stat(resids, klags, X=None, drop_lags_aux=True):
    """Compute Breusch-Pagan-Godfrey LM statistic"""
    T = resids.shape[0]
    # X is lhs matrix from original regression 
    # (should already include a constant)
    # auxilliary regression
    if drop_lags_aux:
        ### drop lags
        resids_lags = lag_mat(resids, klags)[klags:,:]
        if X is not None:
            X = X[klags:,:]
    else:
        ### fill lags with 0
        resids_lags = lag_mat(resids, klags, fill_vals=0)
    T_aux = resids_lags.shape[0]
    lhs = resids_lags[:,0]
    if X is None:
        rhs = resids_lags[:,1:]
    else:
        rhs = np.c_[ X , resids_lags[:,1:]]
    ### OLS
    beta_aux,resids_aux,_ = calc_ols(lhs, rhs, addcon=False)
    ### R-squared
    lhs_demean = lhs - lhs.mean()
    ss_tot = np.dot(lhs_demean, lhs_demean)
    ss_res = np.dot(resids_aux, resids_aux)
    Rsq = 1 - ss_res/ss_tot
    bpg_stat = T_aux*Rsq
    ### Get p-value
    pval = 1 - scipy.stats.chi2.cdf(bpg_stat, klags) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2.html
    return bpg_stat, pval

def calc_llf(resids):
    """Compute (positive) log likelihood given residuals"""
    Nobs = resids.shape[0]
    ### Sum of squared residuals and variance
    ss_res = np.dot(resids, resids)
    sigma2 = ss_res/Nobs
    llf = -(Nobs/2.)*(np.log(sigma2) + np.log(2*np.pi)) - ss_res/(2*sigma2)
    return llf


def calc_aic(resids, k):
    """Compute AIC given residuals and number of parameters."""
    Nobs = resids.shape[0]
    ### (Positive) log-likelihood
    L = calc_llf(resids)
    ### Not scaled by number of obesrvations
    AIC = (-2*L + 2*k)
    return AIC

def calc_bic(resids, k):
    """Compute BIC given residuals and number of parameters."""
    Nobs = resids.shape[0]
    ### (Positive) log-likelihood
    L = calc_llf(resids)
    BIC = (-2*L + k*np.log(Nobs))
    return BIC
```

```python
### test out the function
np.random.seed(200)

### AR(2) process
y_ar2_process = arma_from_roots(ar_roots=[-3.5,1.5], ma_roots=[])
y_ar2 = y_ar2_process.generate_sample(500)
print('AR coefs: ', y_ar2_process.arcoefs)

### Fit an AR(1), AR(2), and AR(3)
nlags = 3
y_lagmat = lag_mat(y_ar2, nlags=nlags)[nlags:,:]
y = y_lagmat[:,0]
Nobs = y.shape[0]

for p in range(3):
    print('\nAR(%d)' % (p+1) )
    x = y_lagmat[:,1:p+2]
    ### Estimate, get residuals and save rhs matrix (with constant)
    beta_hat, resids, X = calc_ols(y,x, addcon=True)
    print('Estimates: ',  beta_hat)
    
    ### test statistics
    dw_stat = durbin_watson_stat(resids)
    print('DW stat: %.3f' % dw_stat)
    
    bpg_stat, pval = breusch_pagan_godfrey_stat(resids, klags=20, X=X,
        drop_lags_aux=False)
    print('BPG stat (pval): %.3f (%.3f)' % (bpg_stat, pval))
    
    ### number of lags and constant
    kparams = p+2
    AIC = calc_aic(resids, kparams)
    BIC = calc_bic(resids, kparams)
    
    print('AIC: %.3f' % AIC)
    print('BIC: %.3f' % BIC)
    
### NOTE: Given a set of candidate models for the data, the preferred model is the one with the minimum AIC/BIC value.   
```

```python
def get_arp_diagnostics(ar_data,p, display=False):
    """Estimate AR(p) model by OLS and compute diagnostics."""
    ### Lags, dropping missing
    ar_data_lags = lag_mat(ar_data, p)[p:,:]
    Nobs = ar_data_lags.shape[0]
    ### LHS and RHS data
    y = ar_data_lags[:,0]
    if p==0:
        ### No fitting
        resids = np.copy(ar_data)
        ### for BPG statistic
        X = None
    else:
        x = ar_data_lags[:,1:]
        ### Estimate, get residuals and save rhs matrix (with constant)
        beta_hat, resids, X = calc_ols(y,x, addcon=True)
        if display: print( 'OLS coefs: ', beta_hat )
    
    ### Diagnostic stats
    dw_stat = durbin_watson_stat(resids)
    bpg_stat, bpg_pval = breusch_pagan_godfrey_stat(resids, klags=20, X=X, 
        drop_lags_aux=False)
    ### Number of lags and constant
    kparams = p+1
    aic = calc_aic(resids, kparams)
    bic = calc_bic(resids, kparams)
    diag_stats = (dw_stat, bpg_stat, bpg_pval, aic, bic)
    if display:
        print( 'DW stat: %.3f' % dw_stat )
        print( 'BPG stat (pval): %.3f (%.3f)' % ( bpg_stat, bpg_pval ) )
        print( 'AIC: %.3f' % aic )
        print( 'BIC: %.3f' % bic )

    return resids, diag_stats
```

### Exercise 8: AR($1$) simulation and OLS estimation

Write a function that simulates and then estimates an AR($1$) process by OLS.

Use the functions you wrote above, or make use of the following statsmodels functions:
* `sm.tsa.arma_generate_sample` (more efficient than our own code; for repeated simulations this will be much faster)
* `sm.OLS`

```python
def sim_est_ar1_ols(ar1_coef, nsample):
    """AR(1) simulation and estimation wrapper"""
    ### Params for sm
    ar = np.r_[1, -ar1_coef] # add zero-lag and negate
    ma = [1]  # add zero-lag
    ar_data = sm.tsa.arma_generate_sample(ar, ma, nsample) # https://www.statsmodels.org/dev/generated/statsmodels.tsa.arima_process.arma_generate_sample.html
    ### estimate (include a constant)
    y = ar_data[1:]
    X = sm.add_constant(ar_data[:-1])
    beta_hat = sm.OLS(y,X).fit().params
    return beta_hat
```

```python
# Test out the function
np.random.seed(10)
print(sim_est_ar1_ols(.5, 100))
print(sim_est_ar1_ols(.5, 1000))
print(sim_est_ar1_ols(.5, 10000))
print(sim_est_ar1_ols(.5, 100000))
```

### Exercise 9: AR($1$) OLS distribution - Part 1

Plot the distribution of the OLS estimator over repeated simulations

```python
def plot_ar1_ols(ar1_coef, nsample, Nsims=100, ax=None):
    ### List containing all regression results
    ar1_results = [sim_est_ar1_ols(ar1_coef, nsample) for _ in range(Nsims)]
    ### Estimated AR(1) coefficient array
    ar1_hats = np.array([b[-1] for b in ar1_results])
    
    ### Set figure
    if ax is None:
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
    ax.hist(ar1_hats, bins=25)
    ### Add the true value
    ymin,ymax = ax.get_ylim()
    ax.plot((ar1_coef,ar1_coef), (ymin,ymax), 'r--')
```

### Exercise 10: AR($1$) OLS distribution - Part 2

Plot the distribution of $\sqrt{T}(\hat{\phi} - \phi)$.

Note: to compare with the theoretical normal PDF, see:
* `mpl.mlab.normpdf`

```python
def plot_ar1_ols_dist(ar1_coef, nsample, Nsims=100, ax=None):
    ### List containing all regression results
    ar1_results = [sim_est_ar1_ols(ar1_coef, nsample) for _ in range(Nsims)]
    ### Estimated AR(1) coefficient array
    ar1_hats = np.array([b[-1] for b in ar1_results])
    ar1_hats = np.sqrt(nsample)*(ar1_hats - ar1_coef) # sqrt(T)* (ar1_hat - ar1_true)
    
    ### Set figure
    if ax is None:
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
    ### Normalized
    ax.hist(ar1_hats, bins=25, density=True) # https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html
    ### Add normal distribution
    xmin,xmax = ax.get_xlim()
    mu = 0
    sigma = np.sqrt(1 - ar1_coef**2)
    x = np.linspace(-3,3, 100)
    ax.plot(x, scipy.stats.norm.pdf(x , mu, sigma))
```

```python
### Test out of the function
np.random.seed(123)
ar1_coef = .9
Nsims = 2000
xlims = [-2.5,2.5]
### Set up figure
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,12))
for idx, nsample in enumerate((50,100,1000,5000)):
    idx_r,idx_c = idx//2, np.mod(idx,2) 
    ax = axes[idx_r,idx_c]
    plot_ar1_ols_dist(ar1_coef, nsample, Nsims=Nsims, ax=ax)
    ### Same x axis
    ax.set_xlim( xlims[0], xlims[1])
    ax.set_title('Sample: %d' % nsample)

fig.tight_layout()
```
